```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn import preprocessing
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.metrics import jaccard_score
from sklearn.metrics import f1_score
from sklearn.metrics import log_loss
import matplotlib.pyplot as plt
from sklearn.metrics import (
    confusion_matrix,
    accuracy_score,
    mean_absolute_error,
    mean_squared_error,
    r2_score
)
import sklearn.metrics as metrics
```

### Importing the Dataset


```python
df = pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ML0101EN-SkillUp/labs/ML-FinalAssignment/Weather_Data.csv')

df.head()
```

### Data Preprocessing


#### Transforming Categorical Variables


First, we need to convert categorical variables to binary variables. We will use pandas `get_dummies()` method for this.


```python
df_sydney_processed = pd.get_dummies(data=df, columns=['RainToday', 'WindGustDir', 'WindDir9am', 'WindDir3pm'])
```

Next, we replace the values of the 'RainTomorrow' column changing them from a categorical column to a binary column. We do not use the `get_dummies` method because we would end up with two columns for 'RainTomorrow' and we do not want, since 'RainTomorrow' is our target.


```python
df_sydney_processed.replace(['No', 'Yes'], [0,1], inplace=True)
```

### Training Data and Test Data


Now, we set our 'features' or x values and our Y or target variable.


```python
df_sydney_processed.drop('Date',axis=1,inplace=True)
```

```python
df_sydney_processed = df_sydney_processed.astype(float)
```

```python
features = df_sydney_processed.drop(columns='RainTomorrow', axis=1)
Y = df_sydney_processed['RainTomorrow']
```

```python
features.shape
```

### Linear Regression


#### Q1) Use the `train_test_split` function to split the `features` and `Y` dataframes with a `test_size` of `0.2` and the `random_state` set to `10`.


```python
# Split data into train and test sets
# Notice th order of the inputs and outputs
X_train, X_test, y_train, y_test = train_test_split(features, Y, test_size=0.2, random_state=10)
```

#### Q2) Create and train a Linear Regression model called LinearReg using the training data (`x_train`, `y_train`).


```python
# Create a linear model object
LinearReg = LinearRegression()

# Fit the model to the training data
LinearReg.fit(X_train, y_train)
```

#### Q3) Now use the `predict` method on the testing data (`x_test`) and save it to the array `predictions`.


```python
predictions_OLS = LinearReg.predict(X_test)
```

#### Q4) Using the `predictions` and the `y_test` dataframe calculate the value for each metric using the appropriate function.


```python
# Calculate the Mean Absolute Error
LinearRegression_MAE = mean_absolute_error(
    y_true=y_test,
    y_pred=predictions_OLS,
)
LinearRegression_MSE = mean_squared_error(
    y_true=y_test,
    y_pred=predictions_OLS,
)
LinearRegression_R2 = r2_score(
    y_true=y_test,
    y_pred=predictions_OLS,
)
```

#### Q5) Show the MAE, MSE, and R2 in a tabular format using data frame for the linear model.


```python
# Make a dict to store the results
report_linear_dict = {
    "Model": ["Linear Regression"],
    "MAE": [round(LinearRegression_MAE, 2)],
    "MSE": [round(LinearRegression_MSE, 2)],
    "R2": [round(LinearRegression_R2, 2)],  
}
# Convert to a dataframe
Report = pd.DataFrame(report_linear_dict)

Report
```

### KNN


#### Q6) Create and train a KNN model called KNN using the training data (`x_train`, `y_train`) with the `n_neighbors` parameter set to `4`.


```python
# Create a model object
KNN = KNeighborsClassifier(n_neighbors=4)
# Fit the training data
KNN.fit(X=X_train, y=y_train)
```

#### Q7) Now use the `predict` method on the testing data (`x_test`) and save it to the array `predictions`.


```python
predictions_KNN = KNN.predict(X=X_test)
```

#### Q8) Using the `predictions` and the `y_test` dataframe calculate the value for each metric using the appropriate function.


```python
KNN_Accuracy_Score = accuracy_score(
    y_true=y_test,
    y_pred=predictions_KNN,
)
KNN_JaccardIndex = jaccard_score(
    y_true=y_test,
    y_pred=predictions_KNN,
)
KNN_F1_Score = f1_score(
    y_true=y_test,
    y_pred=predictions_KNN,
)
```

```python
# Update the report
report_KNN_dict = {
        "Model": "KNN_4",
        "Accuracy Score": round(KNN_Accuracy_Score, 2),
        "Jaccard Index": round(KNN_JaccardIndex, 2),
        "F1 Score": round(KNN_F1_Score, 2),
}
Report_classifier = pd.DataFrame([report_KNN_dict])

print(Report_classifier)
print(Report)
```

### Decision Tree


#### Q9) Create and train a Decision Tree model called Tree using the training data (`x_train`, `y_train`).


```python
# Create a Tree model object
Tree = DecisionTreeClassifier(random_state=10)
# Fit to the training data
Tree.fit(X=X_train, y=y_train)
```

#### Q10) Now use the `predict` method on the testing data (`x_test`) and save it to the array `predictions`.


```python
predictions_tree = Tree.predict(X=X_test)
```

#### Q11) Using the `predictions` and the `y_test` dataframe calculate the value for each metric using the appropriate function.


```python
tree_Accuracy_Score = accuracy_score(
    y_true=y_test,
    y_pred=predictions_tree,
)
tree_JaccardIndex = jaccard_score(
    y_true=y_test,
    y_pred=predictions_tree,
)
tree_F1_Score = f1_score(
    y_true=y_test,
    y_pred=predictions_tree,
)
```

```python
# Update the report
report_tree_dict = {
        "Model": "Decision Tree",
        "Accuracy Score": round(tree_Accuracy_Score, 2),
        "Jaccard Index": round(tree_JaccardIndex, 2),
        "F1 Score": round(tree_F1_Score, 2),
}
Report_classifier = pd.DataFrame(
    [report_KNN_dict, report_tree_dict]
)

print(Report_classifier)
print(Report)
```

### Logistic Regression


#### Q12) Use the `train_test_split` function to split the `features` and `Y` dataframes with a `test_size` of `0.2` and the `random_state` set to `1`.


```python
X_train_new, X_test_new, y_train_new, y_test_new = train_test_split(
    features, Y,
    test_size=0.2,
    random_state=1,
)
```

#### Q13) Create and train a LogisticRegression model called LR using the training data (`x_train`, `y_train`) with the `solver` parameter set to `liblinear`.


```python
# Create a model object
LR = LogisticRegression(solver='liblinear')
# Fit the training data
LR.fit(X=X_train_new, y=y_train_new)
```

#### Q14) Now, use the `predict` method on the testing data (`x_test`) and save it to the array `predictions`.


```python
predictions_logistic = LR.predict(X=X_test_new)
```

#### Q15) Using the `predictions` and the `y_test` dataframe calculate the value for each metric using the appropriate function.


```python
LR_Accuracy_Score = accuracy_score(
    y_true=y_test_new,
    y_pred=predictions_logistic,
)
LR_JaccardIndex = jaccard_score(
    y_true=y_test_new,
    y_pred=predictions_logistic,
)
LR_F1_Score = f1_score(
    y_true=y_test_new,
    y_pred=predictions_logistic,
)
LR_Log_Loss = log_loss(
    y_true=y_test_new,
    y_pred=predictions_logistic,
)
```

```python
report_logistic_dict = {
    "Model": "Logistic",
    "Accuracy Score": LR_Accuracy_Score,
    "Jaccard Index": LR_JaccardIndex,
    "F1 Score": LR_F1_Score,
    "Log Loss": LR_Log_Loss
}
```

### SVM


#### Q16) Create and train a SVM model called SVM using the training data (`x_train`, `y_train`).


```python
SVM = SVC(kernel='linear', random_state=1)
SVM.fit(X_train_new, y_train_new)
```

#### Q17) Now use the `predict` method on the testing data (`x_test`) and save it to the array `predictions`.


```python
predictions_SVM = SVM.predict(X=X_test_new)
```

#### Q18) Using the `predictions` and the `y_test` dataframe calculate the value for each metric using the appropriate function.


```python
SVM_Accuracy_Score = accuracy_score(
    y_true=y_test_new,
    y_pred=predictions_SVM,
)
SVM_JaccardIndex = jaccard_score(
    y_true=y_test_new,
    y_pred=predictions_SVM,
)
SVM_F1_Score = f1_score(
    y_true=y_test_new,
    y_pred=predictions_SVM,
)
```

```python
report_SVM_dict = {
    "Model": "SVM",
    "Accuracy Score": SVM_Accuracy_Score,
    "Jaccard Index": SVM_JaccardIndex,
    "F1 Score": SVM_F1_Score,
}
```
